#### RESPOSTA:

<p align="justify">
A média ou valor esperado de um processo aleatório x(n) é uma função dada por:
    
$$ \overline{x(n)} = E[x(n)] = \int_{-∞}^{∞} xp_x(x;n)  dx $$

A função de autocorrelação é dada pela correlação entre duas variáveis aleatórias nos instantes n1 e n2 : x(n1) e x(n2). Logo temos que:

$$ R_x (n_1,n_2) = \overline{x(n_1)x(n_2)} = \overline{x_1x_2} = \int_{-∞}^{∞}\int_{-∞}^{∞} x_1x_2 Px_1 x_2 (x_1 x_2; n_1 n_2) dx_1 dx_2 $$

Um processo é estacionário no sentido amplo se:

$$\overline{x(n)} = Constante$$
$$R_x (n_1,n_2) = R_x(w), w = n_1 - n_2$$
</p>
