<p align="justify"> <center>Razão de verossimilhança</center>
    
$$L(y) = \frac{P(y|H_1)}{P(y|H_0)}$$

Seguindo o problema dado no referido item, denotamos que com o uso das respectivas pdf, chegamos a seguinte expressão:

$$\frac{L_1}{L_0} = \frac{P(y|H_1)}{P(y|H_0)} \frac{\geq}{\leq} n $$

$$\frac{L_1}{L_0} = \frac{0.25}{0.5} \frac{\geq}{\leq} n $$

</p>
<p align="justify">Tomando o log-verossimilhança, temos que:
$$y \frac{\geq}{\leq} n$$

Determinando as regiões, ficamos com as seguintes expressções para os referidos itens; $n = \frac{1}{4} , n = 1, n = 2$, respectivamente obtemos:

$(1)$ $$  y \frac{\geq}{\leq} \frac{1}{4} $$

$(2)$ $$  y \frac{\geq}{\leq} 1 $$

$(3)$ $$  y \frac{\geq}{\leq} 2 $$
</p>


#### (B) 

<p align="justify">A probabilidade de falso alarme, é definida: $P_F = P_r ({D_1}|{H_0}) = P_{10}$
    
Logo, 

$$P_F = P_r ({D_1}|{H_0}) = P_{10}$$

$$ P_r ({D_1}|{H_0}) =\frac{P_r (D_1 \cap H_0)}{Pr(H_0)}$$

$$P_{10} = \frac{P_r (D_1 \cap H_0)}{Pr(H_0)} $$
</p>

```python
listap1 = []
listap2 = []
listak = []

def degr():
    k = np.arange(-10,10, 0.001)
    p1 = 0
    p2 = 0
    for w in k:
        listak.append(w)
        if (w >= -1 and w<=1):
            p1 = 0.5
        else :
            p1 = 0
        listap1.append(p1)
        if (w >= -2 and w <= 2):
            p2 = 0.25
        else:
            p2 = 0
        listap2.append(p2)
    fig, ax = plt.subplots()
    ax.grid()
    plt.title('GRÁFICO DAS RESPECTIVAS PDF DO SISTEMA')
    ax.plot(listak, listap1, color='red', label='ps(s)')
    ax.plot(listak, listap2, color='black', label='pn(n)')
    ax.legend()
    plt.show()
degr()
```
