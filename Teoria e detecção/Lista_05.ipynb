```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import scipy
import numpy as np
import pandas as pd
import statistics
import math
```

### <center>RECEBIDO<center>

```python
filename = 'Recebido.mat'
# carregando o arquivo matlab
y = loadmat(filename)
```

```python
plt.figure(1)
plt.plot(y['Y'], '--')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
lista = []
lista1 = []
for i in y['Y']:
    a = i 
    lista.append(i)

listay = []
lista10 = lista
x = 0
p = 50
while(x < p):
    listay.append(lista10[x])
    x+=1
    
x1 = np.arange(0, p, 1)

plt.figure(5)
plt.plot(x1, listay, color='black')
plt.grid()
plt.legend()
plt.show()
```

### <center>TRANSMITIDO</center>
