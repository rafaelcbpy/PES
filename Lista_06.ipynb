```python
%matplotlib notebook
from numpy import *
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
 
x = linspace(-1.1, 1.1, 500)
y = linspace(-1.1, 1.1, 500)
x, y = meshgrid(x, y)
z = x + 1j*y
f = z / (z - 1.6)
 
t = linspace(-pi, pi, 500)
xc = 0.5*cos(t)
yc = 0.5*sin(t)
zc = xc + 1j*yc
fc = zc / (zc - 1.6)
 

x1 = cos(t)
y1 = sin(t)
z1 = x1 + 1j*y1
f1 = z1 / (z1 - 1.6)
 

fig = figure()
a = fig.gca(projection='3d')
a.view_init(55, -120)
 
plt.figure(5)
gray()
a.plot(xc, yc, log(abs(fc)+1), 'k--')
a.plot(x1, y1, log(abs(f1)+1), 'k:')
a.grid()
plt.plot(xc, yc, log(abs(fc)+1), 'k--')
plt.plot(x1, y1, log(abs(f1)+1), 'k:')
plt.grid()
a.plot_surface(x, y, log(abs(f)+1.), rstride=16, cstride=16, shade=True)
a.set_xlabel(u'w0')
a.set_ylabel(u'w1')
a.set_zlabel(u'|X(z)|')
a.set_xlim3d([ -1.1, 1.1 ])
a.set_ylim3d([ -1.1, 1.1 ])
a.set_zlim3d([ 0., 5. ])
 
```
